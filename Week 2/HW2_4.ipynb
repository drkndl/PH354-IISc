```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 1 # Initial guess
error = 1E-6
```

```python
def fpi(guess,c):    
    for i in range(1, 1000):
        x_new = 1-np.e**(-c*guess)
        if abs(x_new-guess)<error:
            print("The iteration number: ", i)
            return x_new
        else:
            guess = x_new
```

```python
print(fpi(x,2))
```

```python
arr = np.arange(0,3.01,0.01)
print(arr)
```

```python
xarr = []
for c in arr:
    xarr.append(fpi(x, c))
```

```python
plt.plot(arr, xarr)
plt.grid()
plt.title("Percolation Transition")
```
